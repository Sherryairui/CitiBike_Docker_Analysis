```python
import os
import numpy as np
import pandas as pd
import pylab as pl
import datetime as dt
```

```python
traffic = pd.read_csv("data/Traffic_Volume_Counts__2014-2018_.csv")
```

```python
traffic.head()
```

```python
traffic['Date'] = pd.to_datetime(traffic.Date)
```

```python
traffic_2018 = traffic[traffic.Date.apply(lambda x: x.year) == 2018]
```

```python
trafficVolume = np.mean(traffic_2018).to_frame().iloc[2:].rename(columns = {0: 'Traffic_Volume'})
```

```python
trafficVolume.to_csv('clean_data/traffic_volume.csv', index = False)
```
